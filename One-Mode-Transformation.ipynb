```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

# Generate one-mode network connecting vendors by shared customers

```python
vendors = set();
f = open("result.csv", mode="w");
for key, value in df[df['signingyear'] == 1998].groupby('cid'):
    for v1 in value['vid'].values:
        for v2 in value['vid'].values:
            if v1 != v2:
                key = [v1, v2];
                key.sort()
                if str(key) not in vendors:
                    vendors.add(str(key));
                    f.write(str(v1 + ", " + v2 + "\n"));
f.close()
```

# Generate one-mode network connecting vendors by shared customer industry

```python
vendors = set();
f = open("result.csv", mode="w");
for key, value in df[df['signingyear'] == 1998].groupby('customerindustrycode'):
    for v1 in value['vid'].values:
        for v2 in value['vid'].values:
            if v1 != v2:
                key = [v1, v2];
                key.sort()
                if str(key) not in vendors:
                    vendors.add(str(key));
                    f.write(str(v1 + ", " + v2 + "\n"));
f.close()
```
