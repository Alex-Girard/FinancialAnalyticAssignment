### Define some statistical measures

```python
epsilon = 0.005;

def computeWOE(X, target):
    a = epsilon;
    biv = pd.crosstab(X, target);
    WoE = np.log((1.0*biv[0]/sum(biv[0])+a) / (1.0*biv[1]/sum(biv[1])+a))
    return WoE;

def computeIV(X, target):
    a = epsilon;
    biv = pd.crosstab(X, target);
    IV = sum(((1.0*biv[0]/sum(biv[0])+a) - (1.0*biv[1]/sum(biv[1])+a))*np.log((1.0*biv[0]/sum(biv[0])+a) / (1.0*biv[1]/sum(biv[1])+a)))
    return IV;

def computeGini(y, target):
    roc = ROCData(zip(*[target, y]));
    gini = 2*roc.auc()-1;
    return gini;

def computeKS2(a, b):
    tot_bads=1.0*sum(b)
    tot_goods=1.0*(len(b)-tot_bads)
    elements = sorted(zip(*[a,b]),key= lambda x: x[0])
    ks2 = max_ks2 = goods = bads = perc_bads = perc_goods = 0.0
    for i in range(len(elements)):
        if (int(elements[i][1])==0):
            goods += 1
            perc_goods = goods/tot_goods
        elif (int(elements[i][1])==1):
            bads += 1
            perc_bads = bads/tot_bads
        ks2 = abs(perc_goods - perc_bads)
        if ks2 > max_ks2:
            max_ks2 = ks2
    return max_ks2

def computePSI(actual, expected):
    a = epsilon;
    act = actual.value_counts();
    totalAct = 1.0 * act.values.sum();
    exp = expected.value_counts();
    totalExp = 1.0 * exp.values.sum();
    for actKey in act.keys():
        if actKey not in exp.keys():
            exp = exp.append(pd.Series([a], index=[actKey]));
    for expKey in exp.keys():
        if expKey not in act.keys():
            act = act.append(pd.Series([a], index=[expKey]));
    act = act.values / totalAct;
    exp = exp.values / totalExp;
    try:
        PSI = (act - exp) * np.log(act / exp);
    except:
        print act, exp
    return PSI.sum();
```

### Define some marginal statistical measures

```python
def computeMarginalIV(X, target, biv_pred):
    a = epsilon;
    biv = pd.crosstab(X, target);
    biv = biv - biv_pred.sum();
    IV = sum((((biv[0]+a)/(sum(biv[0])+a))-((biv[1]+a)/(sum(biv[1])+a)))*np.log(((biv[0]+a)/(sum(biv[0])+a)) / ((biv[1]+a)/(sum(biv[1])+a))))
    return IV;
```

# Utility and prediction functions

```python
def evaluateCandidates(df, candidates, trainPred):
    # build a panda dataframe
    values = {
        'id' : [],
        'IV' : [],
        # TODO add more measures like Chi2, hybrids etc
    };
    for i in candidates:
        values['id'].append(i);
        if len(trainPred) == 0:
            values['IV'].append(computeIV(df[i], df[output_var]));
        else:
            values['IV'].append(getMarginalIV(df, i, trainPred));
    return pd.DataFrame(values, index=values['id']);

def orderCandidates(candidateScores):
    avgIV = [];
    for score in candidateScores:
        if len(avgIV) == 0:
            avgIV = score['IV'];
        else:
            avgIV += score['IV'];
    avgIV /= float(len(candidateScores));
    return avgIV.order(ascending=False);

def weighted_choice(choices):
    total = sum(w for c, w in choices);
    r = random.uniform(0, total);
    upto = 0;
    for c, w in choices:
      if upto + w > r:
         return c;
      upto += w;

def predictTree(dftrain, dftest, df, selectedCandidates, paramValue = 1):
    y_train_pred = [];
    y_pred = [];
    y_pred_full = [];
    try:
        X = dftrain[list(selectedCandidates)]
        clf = tree.DecisionTreeClassifier( min_samples_leaf = paramValue);
        clf = clf.fit(X, dftrain[output_var]);
        tree.export_graphviz(clf, out_file='tree.dot', feature_names=list(selectedCandidates))
        y_train_pred = clf.predict_proba(X)[:,1];
        if len(dftest) > 0:
            X = dftest[list(selectedCandidates)]
            y_pred = clf.predict_proba(X)[:,1];
        if len(df) > 0:
            X = df[list(selectedCandidates)]
            y_pred_full = clf.predict_proba(X)[:,1];
    except np.linalg.linalg.LinAlgError as err:
        pass;
    return (y_train_pred, y_pred, y_pred_full), clf;

def predictForest(dftrain, dftest, df, selectedCandidates, paramValue = 10):
    y_train_pred = [];
    y_pred = [];
    y_pred_full = [];
    try:
        X = dftrain[list(selectedCandidates)]
        clf = ensemble.RandomForestClassifier( n_estimators = paramValue);
        clf = clf.fit(X, dftrain[output_var]);
        y_train_pred = clf.predict_proba(X)[:,1];
        if len(dftest) > 0:
            X = dftest[list(selectedCandidates)]
            y_pred = clf.predict_proba(X)[:,1];
        if len(df) > 0:
            X = df[list(selectedCandidates)]
            y_pred_full = clf.predict_proba(X)[:,1];
    except np.linalg.linalg.LinAlgError as err:
        pass;
    return (y_train_pred, y_pred, y_pred_full), clf;

def predictLogit(dftrain, dftest, df, selectedCandidates):
    y_train_pred = [];
    y_pred = [];
    y_pred_full = [];
    resultModel = None;
    try:
        X = dftrain[list(selectedCandidates)]
        #model = sm.Logit(dftrain[output_var],X);
        model = sm.GLM(dftrain[output_var],X, family = sm.families.Binomial());
        resultModel = model.fit()
        y_train_pred = resultModel.predict(X) 
        if len(dftest) > 0:
            X = dftest[list(selectedCandidates)]
            y_pred = resultModel.predict(X)
        if len(df) > 0:
            X = df[list(selectedCandidates)]
            y_pred_full = resultModel.predict(X)
    except np.linalg.linalg.LinAlgError as err:
        pass;
    return (y_train_pred, y_pred, y_pred_full), resultModel;

# redefine this function to use another model
def predict(dftrain, dftest, df, selectedCandidates):
    #sys.stdout.write('.')
    return predictLogit(dftrain, dftest, df, selectedCandidates);

def aggregateResult(v, newV):
    if len(v) == 0:
        return newV;
    elif len(newV) > 0:
        return v + newV;
    else:
        return v;
```

# Build model forward randomly

```python
def buildModelForwardRandomly(df):
    folds = 10;
    trainFolds = 3;
    OOTgini = 0.0;
    OOTks2 = 0.0;
    selectedCandidates = set();
    remainingCandidates = list_all_woe.copy();
    shouldStop = False;
    iteration = 0;
#    dfUnskewed, dfSkewed = overSample(df);
    progress = {
            'iteration' : [],
            'selection' : [],
            'remaining' : [],
            'gini' : [],
            'ks2' : [],
            'tgini' : [],
            'tks2' : [],
            'OOTgini' : [],
            'OOTks2' : [],
            'nextCandidate' : [],
            'improving' : []
        }
    try:
        while remainingCandidates and not shouldStop:
            skf = cross_validation.StratifiedKFold(df[output_var].values, n_folds=trainFolds);
            candidate = np.random.choice(list(remainingCandidates), 1);
            nextCandidate = candidate[0];
            pred_full = [];
            newks2 = 0.0;
            newgini = 0.0;
            for train, test in skf:
                dftrain, dftest = df.ix[train], df.ix[test];
                (y_train_pred, pred_test, y_pred_full), model = predict(dftrain, dftest, df, 
                                                         list(selectedCandidates) + [ nextCandidate ]);
                if len(pred_test) > 0:
                    newks2 += computeKS2(pred_test, dftest[output_var]);
                    newgini += computeGini(pred_test, dftest[output_var]);  
                if len(y_pred_full) > 0:
                    pred_full = aggregateResult(pred_full, y_pred_full);
                else:
                    pred_full = [];
                    break;
            if len(pred_full) > 0:
                pred_full /= float(len(skf));
                newks2 /= float(len(skf));
                newgini /= float(len(skf));
                # compute and keep quality measures
                ks2 = computeKS2(pred_full, df[output_var]);
                gini = computeGini(pred_full, df[output_var]);
                # submit for reality test 
                chance = int(7 * (1 - gini));
                if submissionEnabled and chance > 0 and np.random.randint(0, chance) == 0 and gini > 0.30 and ks2 > 0.30:
                    OOTgini, OOTks2 = submitCandidates(df, skf, selectedCandidates, "_Chris_Group_A_00.csv", 1, predict);    
                progress['iteration'].append(iteration);
                progress['selection'].append(sorted(list(selectedCandidates)));
                progress['remaining'].append(sorted(list(selectedCandidates)));
                progress['tgini'].append(newgini);
                progress['tks2'].append(newks2);
                progress['gini'].append(gini);
                progress['ks2'].append(ks2);
                progress['nextCandidate'].append(nextCandidate);
                progress['OOTgini'].append(OOTgini);
                progress['OOTks2'].append(OOTks2);
                progress['improving'].append(False);
                remainingCandidates.remove(nextCandidate);    
                selectedCandidates.add(nextCandidate);
                        
                if (iteration % 10 == 0):
                    saveProgress(progress);
                displayProgress(progress, iteration);
                iteration += 1;
    finally:
        saveProgress(progress);
    if len(selectedCandidates) > 0:
        skf = cross_validation.StratifiedKFold(df[output_var].values, n_folds=trainFolds);
        OOTgini, OOTks2 = submitCandidates(df, skf, selectedCandidates, "Chris_Group_A_00.csv", 1, predict);  
    return selectedCandidates;
```

```python
def needRemoval(params):
    mu = params.mean();
    sd = params.std();
    minVal = params.min();
    maxVal = params.max();
    result = False;
    if minVal < mu - 3 * sd:
        result = True;
    if maxVal > mu + 3 * sd:
        result = True;
    if result:
        print "*********************"
        print minVal, mu - 3 * sd, mu + 3 * sd, maxVal
        print "*********************"
    return result;

def buildModelStepWise(df):
    folds = 10;
    trainFolds = 3;
    OOTgini = 0.0;
    OOTks2 = 0.0;
    selectedCandidates = set();
    remainingCandidates = list_all_woe.copy();
    removedCandidates = {};
    shouldStop = False;
    trainPred = [];
    latestPredictions = [];
    iteration = 0;
    gini = 0.0;
    ks2 = 0.0;
#    dfUnskewed, dfSkewed = overSample(df);
    progress = {
            'iteration' : [],
            'selection' : [],
            'remaining' : [],
            'gini' : [],
            'ks2' : [],
            'tgini' : [],
            'tks2' : [],
            'OOTgini' : [],
            'OOTks2' : [],
            'nextCandidate' : [],
            'improving' : []
        }
    try:
        while remainingCandidates and not shouldStop:
            skf = cross_validation.StratifiedKFold(df[output_var].values, n_folds=trainFolds);
            candidateScores = [];
            for train, test in skf:
                dftrain = df.ix[train];
                # score candidates
                if len(selectedCandidates) > 0:
                    (trainPred, y_pred, y_pred_full), m = predict(dftrain, [], [], selectedCandidates);
                candidateScores.append(evaluateCandidates(dftrain, remainingCandidates, trainPred));
            # pick best theoritical candidates
            nextCandidates = orderCandidates(candidateScores);
            # pick best actual candidate
            pred_full, params, pvalues, nextCandidate, testks2, testgini, improving = checkNextCandidate(selectedCandidates, nextCandidates, skf, df, removedCandidates);
            if nextCandidate == None: 
                shouldStop = True;
            if len(pred_full) > 0 and nextCandidate != None:
                # compute and keep quality measures
                oldgini = gini;
                oldks2 = ks2;
                ks2 = computeKS2(pred_full, df[output_var]);
                gini = computeGini(pred_full, df[output_var]);
                # submit for reality test 
                chance = int(10 * (1 - gini));
                if submissionEnabled and chance > 0 and np.random.randint(0, chance) == 0 and gini > 0.30 and ks2 > 0.30:
                    OOTgini, OOTks2 = submitCandidates(df, skf, selectedCandidates, "_Chris_Group_A_05.csv", 1, predict);    
                progress['iteration'].append(iteration);
                progress['selection'].append(list(selectedCandidates));
                progress['remaining'].append(list(remainingCandidates));
                progress['tgini'].append(testgini);
                progress['tks2'].append(testks2);
                progress['gini'].append(gini);
                progress['ks2'].append(ks2);
                progress['nextCandidate'].append(nextCandidate);
                progress['OOTgini'].append(OOTgini);
                progress['OOTks2'].append(OOTks2);
                progress['improving'].append(improving);
                remainingCandidates.remove(nextCandidate);    
                selectedCandidates.add(nextCandidate);        
                if (iteration % 10 == 0):
                    saveProgress(progress);
                displayProgress(progress, iteration);
                iteration += 1;
                chance = float(100.0 * abs(gini - oldgini)) + 1.0;
            shot = np.random.randint(0, int(chance));
            #print "chance out of " + str(int(1 + chance)) + ' got ' + str(shot);
            if needRemoval(params) and shot == 0:
                print "Starting to remove variables..."
                shouldStop = False;
                while len(selectedCandidates) > 1 and not shouldStop:
                    skf = cross_validation.StratifiedKFold(df[output_var].values, n_folds=trainFolds);
                    params, pvalues, ks2, gini = evaluatePValues(df, skf, selectedCandidates);
                    pred_full, removalCandidate, testks2, testgini, improving = checkRemovalCandidates(selectedCandidates, pvalues.order(ascending=False), skf, df, ks2, gini);
                    if removalCandidate == None: 
                        shouldStop = True;
                    if len(pred_full) > 0 and removalCandidate != None:
                        # compute and keep quality measures
                        ks2 = computeKS2(pred_full, df[output_var]);
                        gini = computeGini(pred_full, df[output_var]);
                        # submit for reality test 
                        chance = int(7 * (1 - gini));
                        if submissionEnabled and chance > 0 and np.random.randint(0, chance) == 0 and gini > 0.30 and ks2 > 0.30:
                            OOTgini, OOTks2 = submitCandidates(df, skf, selectedCandidates, "_Chris_Group_A.csv", 1);    
                        remainingCandidates.add(removalCandidate);    
                        selectedCandidates.remove(removalCandidate);
                        if (removalCandidate not in removedCandidates):
                            removedCandidates[removalCandidate] = 1;
                        else:
                            removedCandidates[removalCandidate] += 1;
                        progress['iteration'].append(iteration);
                        progress['selection'].append(list(selectedCandidates));
                        progress['remaining'].append(list(remainingCandidates));
                        progress['tgini'].append(testgini);
                        progress['tks2'].append(testks2);
                        progress['gini'].append(gini);
                        progress['ks2'].append(ks2);
                        progress['nextCandidate'].append(removalCandidate);
                        progress['OOTgini'].append(OOTgini);
                        progress['OOTks2'].append(OOTks2);
                        progress['improving'].append(improving);
                        if (iteration % 10 == 0):
                            saveProgress(progress);
                        displayProgress(progress, iteration);
                        iteration += 1;
                shouldStop = False;
            if iteration >= 120:
                shouldStop = True;
    finally:
        saveProgress(progress);
    if len(selectedCandidates) > 0:
        skf = cross_validation.StratifiedKFold(df[output_var].values, n_folds=trainFolds); 
        OOTgini, OOTks2 = submitCandidates(df, skf, selectedCandidates, "Chris_Group_A_05.csv", 1, predict);  
    return selectedCandidates;
```
